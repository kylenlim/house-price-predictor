```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_excel('markhamhouses.xlsx')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
mct = make_column_transformer((OneHotEncoder(), [4]), remainder='passthrough')
X = np.array(mct.fit_transform(X))
mct = make_column_transformer((OneHotEncoder(), [7]), remainder='passthrough')
X = np.array(mct.fit_transform(X))
X = np.asarray(X).astype('float32')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 3)
```

```python
import tensorflow as tf
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))
```

```python
ann.add(tf.keras.layers.Dense(units = 1))
```

```python
ann.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
ann.fit(X_train, y_train, batch_size = 32, epochs = 10000)
```

```python
y_pred = ann.predict(X_test)
```

```python
sum = 0
for i in range(len(y_pred)):
    sum += abs((y_test[i] - y_pred[i])/y_test[i])  
error = 100 * (sum/(len(y_pred)))
print("Relative absolute error : " + str(error))
```

```python
#Type
#[0.0,0.0,1.0] = Townhouse
#[0.0, 1.0, 0.0] = Detached
#[1.0, 0.0, 0.0] = Condo
#----------------------------------
#Style
#[0.0, 0.0, 0.0, 1.0] = Bungalow
#[1.0, 0.0, 0.0, 0.0] = 2-Storey
#[0.0, 0.0, 1.0, 0.0] = Apartment
#[0.0, 1.0, 0.0, 0.0] = 3-Storey
```

```python
def predict_price():
    SquareFeet = float(input("What is the size of the house in square feet?: "))
    Bed = float(input("How many bedrooms does the house have?: "))
    Bath = float(input("How many bathrooms does the house have?: "))
    Type = input("What is the type of the house? (Townhouse, Detached, Condo): ").lower()
    Style = input("What is the style of the house? (Bungalow, 2-Storey, Apartment, 3-Storey): ").lower()
    Taxes = float(input("What is the yearly tax fee of the house?: "))
    valid_house = True
    
    if Type == "townhouse":
        Type = [0.0,0.0,1.0]
    elif Type == "detached":
        Type = [0.0, 1.0, 0.0]
    elif Type == "condo":
        Type = [1.0, 0.0, 0.0]
    else:
        valid_house = False
        
    if Style == "bungalow":
        Style = [0.0, 0.0, 0.0, 1.0]
    elif Style == "2-storey":
        Style = [1.0, 0.0, 0.0, 0.0]
    elif Style == "apartment":
        Style = [0.0, 0.0, 1.0, 0.0]
    elif Style == "3-storey":
        Style = [0.0, 1.0, 0.0, 0.0]
    else:
        valid_house = False
    
    if Bed < 0 or Bath < 0 or SquareFeet < 0 or Taxes < 0:
        valid_house = False
    
    house_info = Type + Style + [SquareFeet, Bed, Bath, Taxes]
    print(house_info)
    prediction = ann.predict([house_info])
    output = prediction[0][0]
    if valid_house == False:
        return print("That is not a valid house, please try again")
    else:
        return print("Your house would cost around $" + "{:.2f}".format(round(output)))
```
